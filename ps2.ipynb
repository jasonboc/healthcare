```python
import pandas as pd
df=pd.read_csv("VTINP16_upd.TXT")
origin=pd.read_csv("origin.csv")
destination=pd.read_csv("destination.csv")
pd.options.display.float_format = '{:.2f}%'.format
```

```python
df51=df[((df['MDC']==5) | (df['MDC']=="5" )) & (df['PPAY']==1)]
df567=df[((df['MDC']==5) | (df['MDC']=="5" )) & ((df['PPAY']==6 )| (df['PPAY']==7 ))]
df81=df[((df['MDC']==8) | (df['MDC']=="8" )) & (df['PPAY']==1)]
df867=df[((df['MDC']==8) | (df['MDC']=="8" )) & ((df['PPAY']==6 )| (df['PPAY']==7 ))]
```

```python
origin=origin.rename(columns = {'HSA Health Service Area ':'hsa','RR Name':'RR origin'})
destination=destination.rename(columns = {'RRName':'RR destination'})
```

```python
new51=pd.merge(df51,origin,on='hsa',how='inner')
new51=pd.merge(new51,destination,on='hnum2',how='inner')
new567=pd.merge(df567,origin,on='hsa',how='inner')
new567=pd.merge(new567,destination,on='hnum2',how='inner')
new81=pd.merge(df81,origin,on='hsa',how='inner')
new81=pd.merge(new81,destination,on='hnum2',how='inner')
new867=pd.merge(df867,origin,on='hsa',how='inner')
new867=pd.merge(new867,destination,on='hnum2',how='inner')
new51=new51[new51['RR origin']!='Z_OutState']
new567=new567[new567['RR origin']!='Z_OutState']
new81=new81[new81['RR origin']!='Z_OutState']
new867=new867[new867['RR origin']!='Z_OutState']
```

```python
result51=new51[['hsa','RR origin','hnum2','RR destination']]
result567=new567[['hsa','RR origin','hnum2','RR destination']]
result81=new81[['hsa','RR origin','hnum2','RR destination']]
result867=new867[['hsa','RR origin','hnum2','RR destination']]
```

```python
od51=pd.crosstab(result51['RR origin'],result51['RR destination'],margins=True)
od51
```
