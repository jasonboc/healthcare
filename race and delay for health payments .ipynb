```python
import pandas as pd
df=pd.read_csv("puf2016.csv")
df
```

```python
df=df[df['ADM_H_MEDSTA']==1]
df
```

```python
df1=df[df['DEM_RACE']==1][["DEM_RACE","ACC_HCDELAY"]]
df1
```

```python
df2=df[df['DEM_RACE']==2][["DEM_RACE","ACC_HCDELAY"]]
df2
```

```python
import numpy as np
import scipy.stats as stats
a1=list(df1.ACC_HCDELAY).count("1")+list(df1.ACC_HCDELAY).count(1)
a2=df1.ACC_HCDELAY.count()-a1
b1=list(df2.ACC_HCDELAY).count("1")+list(df2.ACC_HCDELAY).count(1)
b2=df2.ACC_HCDELAY.count()-b1
obs = np.array([[a1,b1], [a2,b2]])
obs
```

```python
fisher_result = stats.fisher_exact(obs)
p_val = fisher_result[1]
p_val
```
